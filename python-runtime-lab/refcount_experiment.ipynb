```python
import sys

def print_refcount(label, obj):
    print(f"{label}: {sys.getrefcount(obj)}")

a = []
print_refcount("After creation", a)

b = a
print_refcount("After assignment to b", a)

del b
print_refcount("After deleting b", a)
```

```python
d=[]
print(sys.getrefcount(d))
```

```python
def foo(x):
    print_refcount("Inside function", x)

a = []
print_refcount("Before function call", a)
foo(a)
print_refcount("After function call", a)
```

```python
a = []
print_refcount("Initial", a)

container = [a]
print_refcount("After adding to list", a)

del container
print_refcount("After deleting container", a)
```

```python
import gc

class Node:
    def __init__(self):
        self.ref = None

a = Node()
b = Node()

a.ref = b
b.ref = a

del a
del b

print("Collected:", gc.collect())
```

```python
import gc

print("GC enabled:", gc.isenabled())
print("GC thresholds:", gc.get_threshold())
print("GC counts:", gc.get_count())
```

```python
gc.disable()

a = Node()
b = Node()
a.ref = b
b.ref = a

del a
del b

print("Collected:", gc.collect())
```
